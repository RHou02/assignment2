Q1, a): Missing values were handled based on the nature and proportion of missing data.
For numerical columns such as Mileage, Engine, and Power, missing values were replaced using the median to reduce the effect of outliers. Since some numerical columns (Mileage, Engine, and Power) were stored as strings with units, the mode was temporarily used to fill missing values before converting them into numerical format.
For the Seats column, missing values were filled using the mode since it represents categorical-like discrete data.
The New_Price column had more than 85% missing values, so it was dropped from the dataset to avoid unreliable imputation.

```python
# delete New_Price
df = df.drop(columns=['New_Price'], errors='ignore')

df['Mileage'] = df['Mileage'].fillna(df['Mileage'].mode()[0])
df['Engine'] = df['Engine'].fillna(df['Engine'].mode()[0])
df['Power'] = df['Power'].fillna(df['Power'].mode()[0])

df['Seats'] = df['Seats'].fillna(df['Seats'].mode()[0])

# chack again
df.isnull().sum()
```

Q1, 2) : Regular expressions were used to extract only the numeric values from columns such as Mileage, Engine, and Power. This approach ensures that different unit formats (e.g., kmpl and km/kg) are handled consistently before converting the data into numeric format.

```python
df['Power'] = df['Power'].replace('null bhp', None)
```

```python
df['Mileage'] = df['Mileage'].str.extract(r'(\d+\.?\d*)')
df['Engine'] = df['Engine'].str.extract(r'(\d+\.?\d*)')
df['Power'] = df['Power'].str.extract(r'(\d+\.?\d*)')
```

```python
df['Mileage'] = pd.to_numeric(df['Mileage'])
df['Engine'] = pd.to_numeric(df['Engine'])
df['Power'] = pd.to_numeric(df['Power'])
```
